## Write 3 SQL queries that use an INNER join.

##### Query 1. List payments made by customers whose last name starts with an "H" and order by payment amount in descending order

```python
!psql -c "select customer.first_name, customer.last_name, payment.amount from  pagila.customer inner join  pagila.payment on customer.customer_id = payment.customer_id where customer.last_name LIKE 'H%' order by payment.amount desc limit 10"
```

##### Query 2. List customers and their rental history for the film titled "WRONG BEHAVIOR", ordered by customer's last names

```python
# Checking what movie titles there are starting from the bottom and limit to 10
!psql -c "select film.title from pagila.film order by film.title desc limit 10"
```

```python
# List customers and their rental history for the film titled "WRONG BEHAVIOR", ordered by customer's last names
!psql -c "select customer.first_name, customer.last_name, rental.rental_date from pagila.customer inner join pagila.rental on customer.customer_id = rental.customer_id inner join pagila.inventory on rental.inventory_id = inventory.inventory_id inner join pagila.film on film.film_id = inventory.film_id where film.title = 'WRONG BEHAVIOR' order by customer.last_name limit 10"
```

##### Query 3. List customers who rented films in the "Comedy" category, along with the film titles and rental dates. This is ordered by customer ID and rental date and limited to 10. 

```python
# Check what type of movie categories there are
!psql -c "select * from pagila.category limit 10"
```

```python
# List customers who rented films in the "Comedy" category, along with the film titles and rental dates. 
!psql -c "select customer.customer_id, customer.first_name, customer.last_name, film.title, rental.rental_date from pagila.customer inner join pagila.rental on customer.customer_id = rental.customer_id inner join pagila.inventory on rental.inventory_id = inventory.inventory_id inner join pagila.film on inventory.film_id = film.film_id inner join pagila.film_category on film.film_id = film_category.film_id inner join pagila.category on film_category.category_id = category.category_id where category.name = 'Comedy' order by customer.customer_id, rental.rental_date limit 10"
```

## Write 3 SQL queries that use an OUTER join.

##### Query 1. List films and the language they are available in. Display only films in English and order by film title. limit the query by 10

```python
# Check what avaiable languages there are
!psql -c "select * from pagila.language"
```

```python
# List films and the language they are available in. Display only films in English and order by film title.
# Limit the query by 10
!psql -c "select film.title, language.name as film_language from pagila.film left outer join pagila.language on film.language_id = language.language_id where language.language_id = '1' order by film.title limit 10"
```

##### Query 2. List of customers who have made rentals and customers who have not made rentals. This also include rental dates and is order by rental date

```python
!psql -c "select customer.customer_id, customer.first_name, customer.last_name, rental.rental_id, rental.rental_date from pagila.customer left outer join pagila.rental on customer.customer_id = rental.customer_id where rental.rental_id is not null or rental.rental_id is null order by rental.rental_date limit 10"
```

##### Query 3. List of actors and films they have acted in, including actors with no film roles. This is ordered by actor ID and film title.

```python
!psql -c " select actor.actor_id, actor.first_name, actor.last_name, film.title from pagila.actor left outer join pagila.film_actor on actor.actor_id = film_actor.actor_id left outer join pagila.film on film_actor.film_id = film.film_id where film.film_id is not null order by actor.actor_id, film.title limit 10 "
```

## Write 1 SQL query that use a SELF join.

##### Query 1. List actors who share the same first names, but have different last names.

```python
!psql -c " select a1.actor_id, a1.first_name, a1.last_name, a2.actor_id, a2.first_name, a2.last_name from pagila.actor a1 inner join pagila.actor a2 on a1.first_name = a2.first_name and a1.last_name != a2.last_name limit 10"
```
